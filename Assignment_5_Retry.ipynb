```python
#Assignment 5
#Implement an ANN on the Telecom Churn dataset using python/R/etc.
#---------------------------------
#Stage 1. Data Preprocessing
#Step 1.1 Import the Libraries-
#Step 1.2 Load the Dataset
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

df = pd.read_csv('/content/sample_data/Telco-Customer-Churn.csv')

#Step 1.3 Encode Categorical values. Handle other non-numerical data.
from pandas import Series, DataFrame
from pylab import rcParams
import seaborn as ab
import scipy
from scipy.stats import spearmanr
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

#One Hot Encoding for non-binary Categorical Values
ohe = OneHotEncoder(handle_unknown='ignore',sparse_output=False
                    ).set_output(transform='pandas')

ohetransform1=ohe.fit_transform(df[['MultipleLines']])
df=pd.concat([df,ohetransform1],axis=1).drop(columns='MultipleLines')

ohetransform2=ohe.fit_transform(df[['InternetService']])
df=pd.concat([df,ohetransform2],axis=1).drop(columns='InternetService')

ohetransform3=ohe.fit_transform(df[['OnlineSecurity']])
df=pd.concat([df,ohetransform3],axis=1).drop(columns='OnlineSecurity')

ohetransform4=ohe.fit_transform(df[['OnlineBackup']])
df=pd.concat([df,ohetransform4],axis=1).drop(columns='OnlineBackup')

ohetransform5=ohe.fit_transform(df[['DeviceProtection']])
df=pd.concat([df,ohetransform5],axis=1).drop(columns='DeviceProtection')

ohetransform6=ohe.fit_transform(df[['TechSupport']])
df=pd.concat([df,ohetransform6],axis=1).drop(columns='TechSupport')

ohetransform7=ohe.fit_transform(df[['StreamingTV']])
df=pd.concat([df,ohetransform7],axis=1).drop(columns='StreamingTV')

ohetransform8=ohe.fit_transform(df[['StreamingMovies']])
df=pd.concat([df,ohetransform8],axis=1).drop(columns='StreamingMovies')

ohetransform9=ohe.fit_transform(df[['Contract']])
df=pd.concat([df,ohetransform9],axis=1).drop(columns='Contract')

ohetransform10=ohe.fit_transform(df[['PaymentMethod']])
df=pd.concat([df,ohetransform10],axis=1).drop(columns='PaymentMethod')

#Encoding for binary Categorical Values

le_gender = LabelEncoder()
le_SeniorCitizen = LabelEncoder()
le_Partner = LabelEncoder()
le_Dependents = LabelEncoder()
le_PhoneService = LabelEncoder()
le_PaperlessBilling = LabelEncoder()
le_Churn = LabelEncoder()

df['gender_n']=le_gender.fit_transform(df['gender'])
df['SeniorCitizen_n']=le_SeniorCitizen.fit_transform(df['SeniorCitizen'])
df['Partner_n']=le_Partner.fit_transform(df['Partner'])
df['Dependents_n']=le_Dependents.fit_transform(df['Dependents'])
df['PhoneService_n']=le_PhoneService.fit_transform(df['PhoneService'])
df['PaperlessBilling_n']=le_PaperlessBilling.fit_transform(
    df['PaperlessBilling'])
df['Churn_n']=le_Churn.fit_transform(df['Churn'])

df=df.drop(['customerID','gender', 'SeniorCitizen', 'Partner', 'Dependents',
       'PhoneService', 'PaperlessBilling', 'Churn'], axis='columns')


df['TotalCharges']=pd.to_numeric(df['TotalCharges'],errors='coerce')
df['TotalCharges'] = df['TotalCharges'].fillna(0)
df['MonthlyCharges'] = df['MonthlyCharges'].fillna(0)
df.dtypes
```

```python
df.head(3)
```

```python
#Step 1.4 Split Dataset into X and Y.
X = pd.DataFrame(df.iloc[:, 0:40].values)
y = df.iloc[:, 40].values
```

```python
#1.5 Split the X and Y Dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
#1.6 Perform Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

print(X_test)
#After performing feature scaling, all values are normalized (as below)
```

```python
#2. Build Artificial Neural Network
#2.1 Import the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense

#2.2 Initialize the Artificial Neural Network
#2.3 Add the input layer and the first hidden layer

# Create a `Sequential` model and add a Dense layer as the first layer.
model = keras.models.Sequential()
model.add(keras.Input(shape=(40,)))
model.add(keras.layers.Dense(6, activation='relu'))
# Now the model will take as input arrays of shape (None, 40)
#Input dimensions are 40 as we have 40 Independant Variables
# and output arrays of shape (None, 6).

#2.4 Add the second hidden layer
#After the first layer, you don't need to specify input size anymore.
model.add(keras.layers.Dense(6))
model.output_shape

#2.5 Add the output layer
model.add(keras.layers.Dense(1, activation='sigmoid'))
#Sigmoid activation function not only predicts but also provides the probability of
# customer churn.

#3. Train the ANN
#3.1 Compile the ANN
model.compile(optimizer = 'adam', loss = 'binary_crossentropy',
              metrics = ['accuracy'])

#3.2 Fit the ANN to the Training set
model.fit(X_train, y_train, 10, 100)
#The ANN will compare predictions with real results in batches of 10
#The ANN will train in Epochs of 100 to improve its accuracy over time.
```

```python
#Accuracy of 81.42% was achieved in 100 Epochs
#4. Predict the Test Set Results-
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
print(y_pred)
```

```python
#As it is not possible to compare each prediction of Churn one-by-one, we will apply
#confusion matrix

#5. Make the Confusion Matrix
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```
